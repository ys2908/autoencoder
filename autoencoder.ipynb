```python
import numpy as np
#import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
#del data['Unnamed: 32']
#data1=data.loc[:,~data.columns.str.contains('Unnamed:32')]
#print(data1.head(5))
X=data.iloc[0:693, 1:5].values
y=data.iloc[0:693, 0].values
labelencoder_X_1=LabelEncoder()
y=labelencoder_X_1.fit_transform
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from keras.models import Sequential, Model
from keras.layers import Dense, Activation, Input, Lambda, Layer
from keras.regularizers import l2
from keras.optimizers import Adam
#from keras.objectives import binary_crossentropy
#from keras import objectives
from keras.losses import mse, binary_crossentropy
from keras import backend as K
from tensorflow.keras.backend import clear_session
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
from sklearn import preprocessing
np.random.seed(123)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
model = Sequential()

model.add(Dense(10, kernel_regularizer=l2(0.01), activation='relu', input_dim = 4))
model.add(Dense(16, kernel_regularizer=l2(0.01), activation='relu'))
model.add(Dense(8, kernel_regularizer=l2(0.01), activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer=Adam(lr = 0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x = X_train, y = y_train,
          validation_data=[X_test, y_test],
          epochs = 100,
          batch_size=20)
```
